```python
from openai import OpenAI
import pandas as pd
import time

# API key and setup
API_KEY = "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"
client = OpenAI(api_key=API_KEY, base_url="https://api.perplexity.ai")

# Load Excel file
df = pd.read_excel("Group_7_holdings.xlsx")
df["Ownership Status Verified"] = ""
df["Confidence Level (%)"] = 0

# Confidence scoring rules (heuristic based on response text)
def estimate_confidence(response_text):
    text = response_text.lower()
    if "definitely" in text or "certainly" in text:
        return 100
    elif "yes" in text and "not" not in text:
        return 90
    elif "no" in text or "does not belong" in text:
        return 90
    elif "probably" in text:
        return 60
    elif "i don't know" in text or "unknown" in text:
        return 20
    else:
        return 50

# Perplexity verification function
def verify_ownership(brand, holding):
    question = (
        f"Does the brand '{brand}' belong to the holding company '{holding}'? "
        "Respond only with 'yes', 'no', or 'I don't know'. "
        "Do not speculate. If you are not sure, say 'I don't know'."
    )

    try:
        response = client.chat.completions.create(
            model="sonar-pro",
            messages=[
                {
                    "role": "system",
                    "content": (
                        "You are an expert in brand ownership verification. "
                        "For every question, respond only with one word: 'yes', 'no', or 'I don't know'. "
                        "No explanations. No guessing. Strictly factual."
                    ),
                },
                {
                    "role": "user",
                    "content": question
                }
            ]
        )
        reply = response.choices[0].message.content.strip().lower()
        print(f"🔎 Perplexity reply ({brand} → {holding}): {reply}")

        # Determine status
        if "yes" in reply and "not" not in reply:
            status = "OK"
        elif "no" in reply or "not belong" in reply:
            status = "NO"
        elif "i don't know" in reply or "unknown" in reply:
            status = "UNKNOWN"
        else:
            status = "ERROR"

        confidence = estimate_confidence(reply)
        return status, confidence

    except Exception as e:
        print(f"❌ Error for {brand}: {e}")
        return "ERROR", 0

# Process each row
for index, row in df.iterrows():
    brand = str(row["Brand Name"]).strip()
    holding = str(row["Main Holding Name"]).strip()
    
    print(f"🔍 Checking: {brand} → {holding}")
    status, confidence = verify_ownership(brand, holding)
    
    df.at[index, "Ownership Status Verified"] = status
    df.at[index, "Confidence Level (%)"] = confidence

    time.sleep(1.5)  # Respect rate limits

# Export result
df.to_excel("Group_7_holdings_verified.xlsx", index=False)
print("✅ Verification completed. File saved: Group_7_holdings_verified.xlsx")
```

```python
import pandas as pd
import os
from openai import OpenAI
from dotenv import load_dotenv
import re
from datetime import datetime

# 🔐 Charger la clé API Perplexity
load_dotenv()
API_KEY = os.getenv("PERPLEXITY_API_KEY")
client = OpenAI(api_key=API_KEY, base_url="https://api.perplexity.ai")

# 📁 Fichier source
EXCEL_FILE = "Group_7_holdings.xlsx"
SHEET_NAME = "Sheet1"
OUTPUT_FILE = "updated_brand_list.xlsx"

# 📤 Charger les données
df = pd.read_excel(EXCEL_FILE, sheet_name=SHEET_NAME)

# 🗃️ Dictionnaire des marques connues par holding
holding_to_brands = df.groupby("Main Holding Name")["Brand Name"].apply(set).to_dict()

# 🔍 Fonction pour interroger l’API
def get_brands_from_holding(holding):
    prompt = f"""
    Liste-moi toutes les marques les plus connues (filiales ou divisions commerciales) appartenant à la holding suivante : {holding}.
    Réponds uniquement par une liste de noms, séparés par une virgule.
    Ex : Nike, Converse, Jordan
    """
    messages = [
        {"role": "system", "content": "Tu es un expert en stratégie de marque et M&A."},
        {"role": "user", "content": prompt}
    ]
    try:
        response = client.chat.completions.create(
            model="sonar",
            messages=messages,
            extra_body={"search_domain_filter": ["-wikipedia.org"]},
        )
        content = response.choices[0].message.content.strip()
        brands = re.split(r",\s*|\n", content)
        return [b.strip() for b in brands if b.strip()]
    except Exception as e:
        print(f"Erreur API pour {holding} : {e}")
        return []

# 🧾 Liste des nouvelles entrées à ajouter
rows_to_add = []

for holding, known_brands in holding_to_brands.items():
    print(f"\n🔍 Recherche des marques de : {holding}")
    all_brands = get_brands_from_holding(holding)
    new_brands = [b for b in all_brands if b not in known_brands]

    for brand in new_brands:
        rows_to_add.append({
            "Main Holding Name": holding,
            "Brand Name": brand,
            "Statut": "À vérifier",
            "Date de vérification": datetime.now().strftime("%Y-%m-%d %H:%M")
        })

# ➕ Ajouter les nouvelles marques au DataFrame
if rows_to_add:
    new_df = pd.DataFrame(rows_to_add)
    full_df = pd.concat([df, new_df], ignore_index=True)
else:
    full_df = df

# 💾 Sauvegarde finale
full_df.to_excel(OUTPUT_FILE, index=False)
print(f"\n✅ Fichier sauvegardé : {OUTPUT_FILE}")
```
